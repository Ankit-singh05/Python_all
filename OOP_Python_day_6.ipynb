## 📌 Topic 1 — Class

### 📖 Definition
A **class** is a blueprint, template, or prototype used to create objects.
It defines what **attributes** (data/variables) and **methods** (functions/behaviors)
the objects created from it will have.

A class on its own does nothing — it is just the design on paper.
Think of a class like an **architect's blueprint** for a house. The blueprint
is not a house itself — but every house built from it shares the same structure.

### 🔑 Key Points
- Defined using the `class` keyword
- Class names use **PascalCase** convention (e.g., `MyClass`, `BankAccount`)
- A class can have **class variables** (shared by all objects) and **instance variables** (unique per object)
- Everything inside a class is indented
- A class is a **user-defined data type**

### 🧠 Real-World Analogy
| Real World | Python OOP |
|------------|-----------|
| Blueprint of a car | Class |
| An actual car on the road | Object |
| Color, model, speed | Attributes |
| Drive, brake, honk | Methods |

### ⚙️ Syntax
```python
class ClassName:
    class_variable = value        # shared by ALL objects

    def __init__(self, params):   # constructor
        self.instance_var = value # unique to each object

    def method(self):             # behavior
        pass
```

````python
# ── CLASS EXAMPLE ─────────────────────────────────────────────────────────────

class Laptop:
    # Class variable — shared by ALL Laptop objects
    manufacturer_country = "China"
    total_laptops_made   = 0

    def __init__(self, brand, processor, ram_gb, price):
        # Instance variables — unique to EACH Laptop object
        self.brand     = brand
        self.processor = processor
        self.ram_gb    = ram_gb
        self.price     = price
        Laptop.total_laptops_made += 1   # update class variable

    def show_specs(self):
        print(f"Brand     : {self.brand}")
        print(f"Processor : {self.processor}")
        print(f"RAM       : {self.ram_gb} GB")
        print(f"Price     : ₹{self.price:,}")
        print(f"Made in   : {Laptop.manufacturer_country}")
        print("-" * 30)

    def upgrade_ram(self, extra_gb):
        self.ram_gb += extra_gb
        print(f"{self.brand} RAM upgraded to {self.ram_gb} GB")


# ── Using the class ────────────────────────────────────────────────
laptop1 = Laptop("Dell", "Intel i5", 8,  55000)
laptop2 = Laptop("HP",   "AMD Ryzen", 16, 72000)
laptop3 = Laptop("Apple","M2 chip",  16, 120000)

laptop1.show_specs()
laptop2.show_specs()
laptop3.show_specs()

# Class variable is shared — same for all objects
print(f"Total laptops created: {Laptop.total_laptops_made}")

# Upgrade an instance — only affects that object
laptop1.upgrade_ram(8)
print(f"Laptop1 RAM is now {laptop1.ram_gb} GB")
print(f"Laptop2 RAM unchanged: {laptop2.ram_gb} GB")
````

## 📌 Topic 2 — Object

### 📖 Definition
An **object** is a real, concrete instance created from a class.
When you create an object, Python allocates memory for it and
fills in all the attributes using the class blueprint.

An object has:
- **State** — the data stored in its attributes (e.g., name = "Alice")
- **Behavior** — the actions it can perform via methods (e.g., deposit())
- **Identity** — a unique memory address (every object is different)

### 🔑 Key Points
- Created by calling the class like a function: `obj = ClassName(args)`
- Each object has its **own copy** of instance variables
- Multiple objects can be created from one class — all independent
- Objects can be passed to functions, stored in lists, returned from functions
- `id(obj)` gives the memory address — each object has a unique one
- `type(obj)` tells you which class it belongs to

### 🧠 Real-World Analogy
- Class = Recipe for making a cake
- Object = An actual cake baked from that recipe
- You can bake 100 cakes from one recipe — each cake is independent

````python
# ── OBJECT EXAMPLE ────────────────────────────────────────────────────────────

class Student:
    def __init__(self, name, roll_no, marks):
        self.name    = name
        self.roll_no = roll_no
        self.marks   = marks

    def get_grade(self):
        if self.marks >= 90:   return "A+"
        elif self.marks >= 80: return "A"
        elif self.marks >= 70: return "B"
        elif self.marks >= 60: return "C"
        else:                  return "F"

    def show(self):
        print(f"Roll: {self.roll_no} | Name: {self.name:10s} | "
              f"Marks: {self.marks} | Grade: {self.get_grade()}")


# ── Creating multiple INDEPENDENT objects ──────────────────────────
s1 = Student("Alice",   101, 92)
s2 = Student("Bob",     102, 75)
s3 = Student("Charlie", 103, 55)
s4 = Student("Diana",   104, 88)

print("=== Student Report ===")
for student in [s1, s2, s3, s4]:
    student.show()

print()

# Each object is INDEPENDENT — changing s1 doesn't affect s2
s1.marks = 95
print(f"After updating Alice's marks:")
s1.show()
s2.show()   # unchanged

print()

# Objects have unique identity
print(f"id(s1) = {id(s1)}")
print(f"id(s2) = {id(s2)}")
print(f"Are they the same object? {s1 is s2}")

# type() tells which class it belongs to
print(f"type(s1) = {type(s1)}")
print(f"Is s1 a Student? {isinstance(s1, Student)}")

# Objects can be stored in lists and iterated
students = [s1, s2, s3, s4]
avg = sum(s.marks for s in students) / len(students)
print(f"\nClass average marks: {avg:.1f}")
````

## 📌 Topic 3 — Constructor

### 📖 Definition
A **constructor** is a special method that is **automatically called** when
a new object is created from a class. Its job is to **initialize** the
object — setting up its attributes and performing any setup needed.

In Python, the constructor is always named `__init__` (init = initialize).
The double underscores on both sides indicate it is a **dunder** (magic) method.

### 🔑 Key Points
- Always named `__init__` in Python
- Called **automatically** — you never call it manually
- First parameter is always `self` (the new object being created)
- Used to set **initial values** for instance attributes
- A class can only have **one** `__init__` (unlike Java/C++ which allow multiple)
- If you don't define `__init__`, Python uses a default empty one

### 🧠 Types of Constructors
| Type | Description |
|------|-------------|
| **Default Constructor** | No parameters (except self) |
| **Parameterized Constructor** | Accepts parameters to set initial values |

### ⚙️ What happens when you write `obj = MyClass()`?
1. Python creates a new empty object in memory
2. Python calls `__init__` and passes the new object as `self`
3. `__init__` fills in all the attributes
4. The finished object is returned and assigned to `obj`

````python
# ── CONSTRUCTOR EXAMPLE ───────────────────────────────────────────────────────

class Employee:

    # Class variable
    company = "IN2IT Technologies"

    # ── Parameterized Constructor ──────────────────────────────────
    def __init__(self, name, emp_id, department, salary):
        print(f"  ✅ Constructor called — creating object for {name}")
        # Instance variables set here
        self.name       = name
        self.emp_id     = emp_id
        self.department = department
        self.salary     = salary
        self.is_active  = True          # default value
        self.projects   = []            # empty list for each employee

    def assign_project(self, project_name):
        self.projects.append(project_name)

    def give_raise(self, percent):
        increase = self.salary * (percent / 100)
        self.salary += increase
        print(f"  {self.name} got {percent}% raise → New salary: ₹{self.salary:,.0f}")

    def show(self):
        print(f"  ID: {self.emp_id} | {self.name} | {self.department} | "
              f"₹{self.salary:,} | Projects: {self.projects}")


print("=== Creating Employee Objects ===")
print("(Watch constructor called message for each)")
print()

e1 = Employee("Arjun",  "E001", "Engineering", 75000)
e2 = Employee("Priya",  "E002", "Marketing",   60000)
e3 = Employee("Rahul",  "E003", "HR",          50000)

print()
print("=== Initial State ===")
e1.show()
e2.show()
e3.show()

print()
print("=== After Operations ===")
e1.assign_project("Customer Portal")
e1.assign_project("AI Dashboard")
e1.give_raise(10)
e2.assign_project("Brand Campaign")
e1.show()
e2.show()
````

````python
# ── Default Constructor vs Parameterized Constructor ──────────────────────────

class DefaultExample:
    def __init__(self):              # Default — no parameters
        self.value = 0
        self.items = []
        print("Default constructor called — value set to 0")

class WithDefaults:
    def __init__(self, x=10, y=20): # Parameters WITH default values
        self.x = x
        self.y = y
        print(f"Constructor called with x={self.x}, y={self.y}")

print("--- Default Constructor ---")
d = DefaultExample()
print(f"d.value = {d.value}")

print()
print("--- With Default Values ---")
w1 = WithDefaults()          # uses defaults: x=10, y=20
w2 = WithDefaults(5)         # x=5, y=20 (default)
w3 = WithDefaults(5, 15)     # x=5, y=15 (both provided)
````

## 📌 Topic 4 — Destructor

### 📖 Definition
A **destructor** is a special method that is **automatically called** when
an object is about to be **destroyed** and its memory is about to be freed.
In Python, the destructor is the `__del__` method.

Python uses **automatic garbage collection** — it tracks how many variables
point to an object (reference count). When the count reaches zero, the
object is destroyed and `__del__` is called.

### 🔑 Key Points
- Method name: `__del__`
- Called automatically when object goes out of scope or is explicitly deleted with `del`
- Used for **cleanup** tasks: closing files, closing DB connections, releasing locks
- Python's garbage collector usually handles memory — destructor handles **external resources**
- Not called immediately in all cases (depends on garbage collector)
- Opposite of constructor: constructor = setup, destructor = cleanup

### 🧠 When is `__del__` called?
- When you use `del obj` explicitly
- When the function ends and local objects go out of scope
- When the program exits

### ⚙️ Constructor vs Destructor
| | Constructor `__init__` | Destructor `__del__` |
|--|------------------------|----------------------|
| When called | Object **created** | Object **destroyed** |
| Purpose | Setup / initialize | Cleanup / release |
| Called by | `ClassName()` | `del obj` or garbage collector |

````python
# ── DESTRUCTOR EXAMPLE ────────────────────────────────────────────────────────

class FileHandler:

    def __init__(self, filename, mode="r"):
        self.filename = filename
        self.mode     = mode
        # Simulate opening a file
        print(f"  📂 [CONSTRUCTOR] File '{self.filename}' opened in '{self.mode}' mode.")

    def read(self):
        print(f"  📖 Reading data from {self.filename}...")

    def write(self, data):
        print(f"  ✏️  Writing '{data}' to {self.filename}...")

    def __del__(self):             # Destructor
        # Simulate closing the file — always done when object is destroyed
        print(f"  🔒 [DESTRUCTOR] File '{self.filename}' closed and resources released.")


print("=== File Operations Demo ===")
print()

# Object created — constructor runs
fh1 = FileHandler("report.txt", "w")
fh1.write("Q1 Sales Data")
fh1.write("Q2 Sales Data")

print()
fh2 = FileHandler("config.json", "r")
fh2.read()

print()
print("--- Explicitly deleting fh1 ---")
del fh1   # ← destructor runs NOW for fh1

print()
print("--- fh2 still alive ---")
fh2.read()

print()
print("--- Program ending — fh2 will be destroyed ---")
# fh2 destructor will run when the cell ends / object goes out of scope
````

````python
# ── Destructor with scope demo ────────────────────────────────────────────────

class DatabaseConnection:

    def __init__(self, db_name):
        self.db_name = db_name
        print(f"  ✅ Connected to '{self.db_name}'")

    def execute(self, query):
        print(f"  🔍 Executing on {self.db_name}: {query}")

    def __del__(self):
        print(f"  🔌 Disconnected from '{self.db_name}'")


def do_database_work():
    print("--- Function starts ---")
    conn = DatabaseConnection("employees_db")   # created inside function
    conn.execute("SELECT * FROM employees")
    conn.execute("UPDATE salary SET amount = 50000")
    print("--- Function ends — conn goes out of scope ---")
    # conn is a local variable — destroyed when function ends → __del__ called here!

do_database_work()
print("--- Back in main code ---")
````

## 📌 Topic 5 — Encapsulation

### 📖 Definition
**Encapsulation** is the OOP principle of **bundling data (attributes) and
methods together** inside a class, AND **restricting direct access** to
some of that data from outside the class.

The idea is: **hide what should be hidden, expose what should be exposed.**
You create a protective boundary around the internal state of an object
and only allow interaction through controlled methods.

### 🔑 Key Points
- One of the **4 pillars** of OOP
- Achieved using access modifiers (naming conventions in Python)
- Prevents **accidental corruption** of data
- Forces all changes to go through **validation logic**
- Makes code easier to **maintain and debug**

### 🔐 Access Levels in Python
| Notation | Level | Meaning |
|----------|-------|---------|
| `name` | Public | Accessible from anywhere |
| `_name` | Protected | Convention — "internal use, handle carefully" |
| `__name` | Private | Name mangling applied — hard to access from outside |

### 🧠 Real-World Analogy
Think of a **capsule** (medicine). The outer shell hides the ingredients inside.
You can't directly access the chemicals — you take the whole capsule.
That protective shell is encapsulation.

### ⚙️ Why is it important?
Without encapsulation:
```python
acc.__balance = 999999  # anyone can directly corrupt data!
```
With encapsulation:
```python
acc.deposit(999999)     # must go through validation method
```

````python
# ── ENCAPSULATION EXAMPLE ─────────────────────────────────────────────────────

class BankAccount:

    bank_name = "IN2IT Bank"   # public class variable

    def __init__(self, owner, account_no, balance):
        self.owner       = owner          # public
        self._account_no = account_no     # protected (convention)
        self.__balance   = balance        # private (name mangled → _BankAccount__balance)
        self.__pin       = "0000"         # private
        self.__transactions = []          # private transaction history

    # ── Public methods — the ONLY way to interact with balance ────
    def deposit(self, amount):
        if not isinstance(amount, (int, float)):
            print("  ❌ Invalid amount type.")
            return
        if amount <= 0:
            print("  ❌ Deposit amount must be positive.")
            return
        self.__balance += amount
        self.__transactions.append(f"CREDIT: +₹{amount:,}")
        print(f"  ✅ Deposited ₹{amount:,} | New Balance: ₹{self.__balance:,}")

    def withdraw(self, amount, pin):
        if pin != self.__pin:
            print("  ❌ Wrong PIN! Transaction declined.")
            return
        if amount <= 0:
            print("  ❌ Withdrawal amount must be positive.")
            return
        if amount > self.__balance:
            print(f"  ❌ Insufficient funds! Balance: ₹{self.__balance:,}")
            return
        self.__balance -= amount
        self.__transactions.append(f"DEBIT : -₹{amount:,}")
        print(f"  ✅ Withdrawn ₹{amount:,} | Remaining: ₹{self.__balance:,}")

    def check_balance(self):             # controlled read
        print(f"  💰 Balance: ₹{self.__balance:,}")

    def change_pin(self, old_pin, new_pin):
        if old_pin != self.__pin:
            print("  ❌ Old PIN incorrect.")
            return
        if len(str(new_pin)) != 4:
            print("  ❌ PIN must be exactly 4 digits.")
            return
        self.__pin = str(new_pin)
        print("  ✅ PIN changed successfully.")

    def show_statement(self):
        print(f"  ── Account Statement for {self.owner} ──")
        for txn in self.__transactions:
            print(f"    {txn}")
        print(f"    Current Balance: ₹{self.__balance:,}")


print("=== Bank Account Demo ===")
acc = BankAccount("Alice", "ACC-001", 10000)

acc.deposit(5000)
acc.deposit(-200)        # validation catches this
acc.withdraw(3000, "0000")
acc.withdraw(3000, "1111")  # wrong PIN
acc.withdraw(50000, "0000") # insufficient funds
acc.check_balance()
acc.change_pin("0000", "4321")
acc.withdraw(1000, "4321")  # new PIN works

print()
acc.show_statement()

print()
print("=== Access Level Demo ===")
print(f"Public   owner: {acc.owner}")           # ✅ accessible
print(f"Protected _account_no: {acc._account_no}")  # ⚠️ accessible but shouldn't
# print(acc.__balance)  # ❌ AttributeError!
print(f"Private via mangling: {acc._BankAccount__balance}")  # technically works but BAD PRACTICE
````

## 📌 Topic 6 — Getter and Setter

### 📖 Definition
- A **Getter** is a method that **reads** (returns) the value of a private attribute
- A **Setter** is a method that **writes** (updates) the value of a private attribute, usually with validation

Together, getters and setters are the **controlled gateway** between private
data and the outside world. They are the practical implementation of encapsulation.

### 🔑 Key Points
- In Python, the cleanest approach uses the **`@property` decorator**
- `@property` makes a getter look like a regular attribute access (no parentheses!)
- `@attribute_name.setter` defines the corresponding setter
- Validation logic lives in the setter — bad data is caught before it reaches the attribute
- Without setter defined, the property becomes **read-only**

### 🧠 Two Approaches in Python

**Approach 1 — Regular methods (old style):**
```python
def get_age(self):    return self.__age
def set_age(self, v): self.__age = v
# Usage: obj.get_age()  / obj.set_age(25)
```

**Approach 2 — @property decorator (Pythonic/modern):**
```python
@property
def age(self): return self.__age

@age.setter
def age(self, v): self.__age = v
# Usage: obj.age  / obj.age = 25  ← looks like attribute access!
```

The `@property` approach is **preferred** because it looks clean and natural.

````python
# ── GETTER & SETTER EXAMPLE ───────────────────────────────────────────────────

class Person:

    def __init__(self, name, age, email):
        self.name   = name           # public (no validation needed)
        self.__age   = age           # private
        self.__email = email         # private

    # ── GETTER for age ────────────────────────────────────────────
    @property
    def age(self):
        return self.__age

    # ── SETTER for age with validation ────────────────────────────
    @age.setter
    def age(self, value):
        if not isinstance(value, int):
            raise TypeError("Age must be an integer.")
        if value < 0:
            raise ValueError("Age cannot be negative.")
        if value > 120:
            raise ValueError("Age cannot exceed 120.")
        self.__age = value

    # ── GETTER for email ──────────────────────────────────────────
    @property
    def email(self):
        # Mask part of email for privacy
        user, domain = self.__email.split("@")
        masked = user[:2] + "***@" + domain
        return masked

    # ── SETTER for email with validation ──────────────────────────
    @email.setter
    def email(self, value):
        if "@" not in value or "." not in value:
            raise ValueError("Invalid email format.")
        self.__email = value

    # ── Read-only property (no setter defined) ────────────────────
    @property
    def is_adult(self):
        return self.__age >= 18     # computed, not stored

    def show(self):
        print(f"  Name    : {self.name}")
        print(f"  Age     : {self.age}")        # uses getter
        print(f"  Email   : {self.email}")      # uses getter (masked)
        print(f"  Adult?  : {self.is_adult}")   # computed property


print("=== Person Object ===")
p = Person("Alice", 25, "alice.sharma@gmail.com")
p.show()

print()
print("=== Updating with Setter ===")
p.age = 30
print(f"  Age updated to: {p.age}")

p.email = "alice.new@company.com"
print(f"  Email updated: {p.email}")

print()
print("=== Setter Validation in Action ===")

try:
    p.age = -5
except ValueError as e:
    print(f"  ❌ ValueError: {e}")

try:
    p.age = "twenty"
except TypeError as e:
    print(f"  ❌ TypeError: {e}")

try:
    p.email = "not-an-email"
except ValueError as e:
    print(f"  ❌ ValueError: {e}")

try:
    p.is_adult = False   # read-only property — no setter!
except AttributeError as e:
    print(f"  ❌ AttributeError: {e}")
````

````python
# ── Old style getter/setter (for comparison) ──────────────────────────────────

class Temperature:

    def __init__(self, celsius):
        self.__celsius = celsius

    # Old style — regular methods
    def get_celsius(self):
        return self.__celsius

    def set_celsius(self, value):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self.__celsius = value

    # @property style — modern Pythonic way
    @property
    def fahrenheit(self):               # read-only — computed from celsius
        return (self.__celsius * 9/5) + 32

    @property
    def kelvin(self):                   # read-only — computed
        return self.__celsius + 273.15


t = Temperature(100)
print(f"Celsius    : {t.get_celsius()} °C")   # old style call
print(f"Fahrenheit : {t.fahrenheit} °F")      # property — no ()
print(f"Kelvin     : {t.kelvin} K")           # property — no ()

t.set_celsius(0)
print(f"\nAfter update → Celsius: {t.get_celsius()} °C | Fahrenheit: {t.fahrenheit} °F")
````

## 📌 Topic 7 — Inheritance

### 📖 Definition
**Inheritance** is the mechanism by which a **child (subclass)** automatically
receives all the **attributes and methods** of its **parent (superclass)**.
The child can then add new things or modify existing ones.

This is the OOP principle of **"Don't Repeat Yourself" (DRY)** — write
common code once in the parent, and all children get it for free.

### 🔑 Key Points
- Achieved using `class Child(Parent):`
- `super()` is used to call the parent's methods
- Child inherits: all public and protected attributes/methods
- Child does NOT directly inherit: private attributes (though they exist)
- `isinstance(obj, Parent)` returns True for child objects too
- `issubclass(Child, Parent)` returns True

### 🧠 Types of Inheritance
| Type | Description | Example |
|------|-------------|---------|
| **Single** | One parent, one child | `class Dog(Animal)` |
| **Multiple** | Two parents, one child | `class C(A, B)` |
| **Multilevel** | Chain of inheritance | `A → B → C` |
| **Hierarchical** | One parent, many children | `Animal → Dog, Cat, Bird` |
| **Hybrid** | Combination of above types | Mix of multiple + multilevel |

### ⚙️ The `super()` function
`super()` gives access to the parent class. Use it to:
- Call the parent's `__init__` from the child's `__init__`
- Call the parent's method from an overridden method

### 🧠 Real-World Analogy
A `SavingsAccount` IS-A `BankAccount`. It has everything a BankAccount has,
plus its own interest calculation. This IS-A relationship = Inheritance.

````python
# ── INHERITANCE EXAMPLE ───────────────────────────────────────────────────────

# ── PARENT CLASS ──────────────────────────────────────────────────
class Animal:
    def __init__(self, name, age, weight_kg):
        self.name      = name
        self.age       = age
        self.weight_kg = weight_kg

    def eat(self):
        print(f"  🍽️  {self.name} is eating.")

    def sleep(self):
        print(f"  😴 {self.name} is sleeping.")

    def breathe(self):
        print(f"  💨 {self.name} is breathing.")

    def info(self):
        print(f"  🐾 Name: {self.name} | Age: {self.age}y | Weight: {self.weight_kg}kg")


# ── CHILD CLASS 1 — Single Inheritance ────────────────────────────
class Dog(Animal):
    def __init__(self, name, age, weight_kg, breed):
        super().__init__(name, age, weight_kg)  # call parent constructor
        self.breed = breed                      # Dog's own attribute

    def bark(self):
        print(f"  🐕 {self.name} says: WOOF WOOF!")

    def fetch(self, item):
        print(f"  🎾 {self.name} fetches the {item}!")

    def info(self):             # OVERRIDE parent's info
        super().info()          # still show parent info
        print(f"     Breed: {self.breed}")


# ── CHILD CLASS 2 ──────────────────────────────────────────────────
class Bird(Animal):
    def __init__(self, name, age, weight_kg, can_fly):
        super().__init__(name, age, weight_kg)
        self.can_fly = can_fly

    def chirp(self):
        print(f"  🐦 {self.name} says: Tweet tweet!")

    def fly(self):
        if self.can_fly:
            print(f"  ✈️  {self.name} is flying high!")
        else:
            print(f"  🚶 {self.name} cannot fly — walking instead.")

    def info(self):
        super().info()
        print(f"     Can fly: {'Yes' if self.can_fly else 'No'}")


# ── GRANDCHILD — Multilevel Inheritance ───────────────────────────
class GuideDog(Dog):           # GuideDog → Dog → Animal
    def __init__(self, name, age, weight_kg, breed, owner_name):
        super().__init__(name, age, weight_kg, breed)
        self.owner_name = owner_name

    def guide(self):
        print(f"  🦮 {self.name} is guiding {self.owner_name} safely.")


print("=== Animal Kingdom ===")
dog  = Dog("Bruno", 3, 25, "Labrador")
bird = Bird("Tweety", 1, 0.5, True)
penguin = Bird("Pingu", 2, 10, False)
guide = GuideDog("Rex", 4, 28, "German Shepherd", "Mr. Sharma")

print("\n-- Dog --")
dog.info()
dog.eat()           # inherited from Animal
dog.bark()          # Dog's own method
dog.fetch("ball")   # Dog's own method

print("\n-- Bird --")
bird.info()
bird.chirp()
bird.fly()

print("\n-- Penguin --")
penguin.info()
penguin.fly()

print("\n-- Guide Dog (multilevel) --")
guide.info()        # shows breed too — inherited from Dog
guide.guide()       # GuideDog's own method
guide.bark()        # inherited from Dog
guide.eat()         # inherited from Animal

print("\n=== isinstance checks ===")
print(f"guide is GuideDog : {isinstance(guide, GuideDog)}")
print(f"guide is Dog      : {isinstance(guide, Dog)}")
print(f"guide is Animal   : {isinstance(guide, Animal)}")
print(f"guide is Bird     : {isinstance(guide, Bird)}")
````

````python
# ── Multiple Inheritance ──────────────────────────────────────────────────────

class Flyable:
    def fly(self):
        print(f"  ✈️  {self.name} can fly!")

class Swimmable:
    def swim(self):
        print(f"  🏊 {self.name} can swim!")

class Runneable:
    def run(self):
        print(f"  🏃 {self.name} can run!")

class Duck(Animal, Flyable, Swimmable, Runneable):   # Multiple Inheritance
    def __init__(self, name, age, weight_kg):
        super().__init__(name, age, weight_kg)

    def quack(self):
        print(f"  🦆 {self.name} says: QUACK!")


donald = Duck("Donald", 2, 1.5)
donald.info()    # from Animal
donald.fly()     # from Flyable
donald.swim()    # from Swimmable
donald.run()     # from Runneable
donald.quack()   # Duck's own

# MRO — Method Resolution Order (how Python decides which method to use)
print("\nMethod Resolution Order for Duck:")
for cls in Duck.__mro__:
    print(f"  {cls}")
````

## 📌 Topic 8 — Method Overriding

### 📖 Definition
**Method Overriding** is when a **child class** provides its own specific
implementation of a method that is **already defined in the parent class**.
The child's version completely **replaces** the parent's version when called
on a child object.

The method in the child must have the **same name** and the **same parameters**
as the method in the parent.

### 🔑 Key Points
- Same method name in both parent and child
- Child's version is used when called on child object
- Parent's version is used when called on parent object
- You can still call parent's version using `super().method_name()`
- This is a form of **runtime polymorphism**
- Allows each subclass to have its own **specialized behavior**

### 🧠 Why Override?
The parent defines a **general** version. The child provides a **specific** version.
Example: `Animal.speak()` says "..." but `Dog.speak()` says "Woof!"

### ⚙️ Key Difference from Overloading
| | Method Overriding | Method Overloading |
|--|-------------------|--------------------|
| Where | Parent vs Child class | Same class |
| Signature | Same name, same params | Same name, different params |
| Purpose | Specialize inherited behavior | Handle different param types |

````python
# ── METHOD OVERRIDING EXAMPLE ─────────────────────────────────────────────────

class Payment:
    def __init__(self, amount, currency="INR"):
        self.amount   = amount
        self.currency = currency

    def process(self):                   # Parent version — generic
        print(f"  💳 Processing payment of {self.currency} {self.amount:,}")

    def validate(self):
        if self.amount <= 0:
            raise ValueError("Amount must be positive")
        return True

    def receipt(self):
        print(f"  🧾 Receipt: {self.currency} {self.amount:,} paid.")


class CreditCardPayment(Payment):
    def __init__(self, amount, card_number, cvv):
        super().__init__(amount)
        self.card_number = card_number
        self.cvv         = cvv

    def process(self):                   # OVERRIDE — credit card specific steps
        print(f"  💳 Credit Card: **** **** **** {str(self.card_number)[-4:]}")
        print(f"  🔐 Verifying CVV...")
        print(f"  🏦 Contacting card network...")
        print(f"  ✅ Credit card payment of ₹{self.amount:,} approved!")

    def receipt(self):                   # OVERRIDE
        super().receipt()                # call parent's receipt too
        print(f"  📧 E-receipt sent to registered email.")


class UPIPayment(Payment):
    def __init__(self, amount, upi_id):
        super().__init__(amount)
        self.upi_id = upi_id

    def process(self):                   # OVERRIDE — UPI specific
        print(f"  📱 UPI ID: {self.upi_id}")
        print(f"  🔔 Sending payment request to your UPI app...")
        print(f"  ✅ UPI payment of ₹{self.amount:,} confirmed!")

    def receipt(self):                   # OVERRIDE
        super().receipt()
        print(f"  📱 SMS confirmation sent to registered mobile.")


class CashPayment(Payment):
    def __init__(self, amount, tendered):
        super().__init__(amount)
        self.tendered = tendered

    def process(self):                   # OVERRIDE — cash specific
        change = self.tendered - self.amount
        print(f"  💵 Cash tendered: ₹{self.tendered:,}")
        print(f"  💰 Amount: ₹{self.amount:,}")
        print(f"  🪙 Change returned: ₹{change:,}")
        print(f"  ✅ Cash payment of ₹{self.amount:,} received!")


payments = [
    CreditCardPayment(5000, 4532123456789012, 123),
    UPIPayment(1200, "alice@upi"),
    CashPayment(800, 1000),
]

for p in payments:
    print(f"\n── {p.__class__.__name__} ──")
    p.process()    # each calls its OWN overridden version
    p.receipt()
````

## 📌 Topic 9 — Polymorphism

### 📖 Definition
**Polymorphism** means **"many forms"** (Greek: poly = many, morph = form).
In OOP, it means the **same method name** can produce **different behavior**
depending on which object is calling it.

You call the same method on different objects, and each responds in its
own way — without you needing to know which specific type of object it is.

### 🔑 Key Points
- Same interface, different implementations
- Makes code **flexible and extensible** — add new classes without changing existing code
- Python uses **Duck Typing**: "If it has the method, call it — regardless of type"
- Two main types in Python:
  - **Method Overriding** (runtime polymorphism)
  - **Duck Typing** (Python's natural polymorphism)

### 🧠 Duck Typing
> *"If it walks like a duck and quacks like a duck, it must be a duck."*

Python doesn't check if an object IS a certain type.
It only checks if the object **HAS the method you want to call**.
If it does — call it. This is extremely powerful.

### 🧠 Real-World Analogy
The word **"speak"** is polymorphic:
- Tell a **dog** to speak → barks
- Tell a **cat** to speak → meows
- Tell a **person** to speak → talks
Same instruction. Different result based on who receives it.

````python
# ── POLYMORPHISM EXAMPLE ──────────────────────────────────────────────────────

import math

class Shape:
    def __init__(self, color="white"):
        self.color = color

    def area(self):
        raise NotImplementedError("Subclass must implement area()")

    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter()")

    def describe(self):
        # This method is POLYMORPHIC — it calls area() and perimeter()
        # which behave differently for each shape
        # describe() doesn't know or care which shape it's dealing with!
        name = self.__class__.__name__
        print(f"  {self.color.capitalize()} {name:12s} | "
              f"Area: {self.area():8.2f} | "
              f"Perimeter: {self.perimeter():.2f}")


class Circle(Shape):
    def __init__(self, radius, color="red"):
        super().__init__(color)
        self.radius = radius

    def area(self):                   # polymorphic method
        return math.pi * self.radius ** 2

    def perimeter(self):              # polymorphic method
        return 2 * math.pi * self.radius


class Rectangle(Shape):
    def __init__(self, width, height, color="blue"):
        super().__init__(color)
        self.width  = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


class Triangle(Shape):
    def __init__(self, a, b, c, color="green"):
        super().__init__(color)
        self.a, self.b, self.c = a, b, c

    def area(self):
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s-self.a) * (s-self.b) * (s-self.c))

    def perimeter(self):
        return self.a + self.b + self.c


class Square(Rectangle):
    def __init__(self, side, color="yellow"):
        super().__init__(side, side, color)


# ── POLYMORPHISM IN ACTION ─────────────────────────────────────────
shapes = [
    Circle(7),
    Rectangle(5, 10),
    Triangle(3, 4, 5),
    Square(6),
    Circle(3, "purple"),
]

print("=== Shape Descriptions (Polymorphism) ===")
for shape in shapes:
    shape.describe()      # same call → different result every time!

total_area = sum(shape.area() for shape in shapes)
print(f"\n  Total area of all shapes: {total_area:.2f}")

print()
print("=== Duck Typing Demo ===")

# This class does NOT inherit from Shape at all!
class LandPlot:
    def __init__(self, length, breadth):
        self.length   = length
        self.breadth  = breadth

    def area(self):           # has .area() — that's all Python cares about!
        return self.length * self.breadth

    def perimeter(self):
        return 2 * (self.length + self.breadth)

    def describe(self):
        print(f"  LandPlot           | Area: {self.area():8.2f} | Perimeter: {self.perimeter():.2f}")


# Mix Shape objects with LandPlot — Python doesn't care about type!
mixed = shapes + [LandPlot(20, 30)]
print("Including non-Shape LandPlot:")
for item in mixed:
    item.describe()  # works as long as describe() exists!
````

## 📌 Topic 10 — Method Overloading

### 📖 Definition
**Method Overloading** means having **multiple methods with the same name**
but different **number or types of parameters** in the same class.
Each version handles a different case.

### 🔑 Key Points — Python vs Other Languages
In Java/C++, you can define multiple methods with the same name:
```java
void add(int a, int b) { ... }
void add(int a, int b, int c) { ... }  // valid in Java
```

**In Python, this does NOT work** — if you define the same method twice,
the second one simply **replaces** the first. Python only keeps the last definition.

Python achieves the same effect through:
1. **Default parameter values** — `def add(a, b=0, c=0)`
2. **`*args`** — accept any number of positional arguments
3. **`**kwargs`** — accept any number of keyword arguments
4. **Type checking inside the method** — inspect the arguments and branch

### ⚙️ Key Difference from Overriding
| | Method Overloading | Method Overriding |
|--|--------------------|--------------------|
| Where | **Same class** | Parent and Child class |
| Name | Same | Same |
| Parameters | Different | Same |
| Purpose | Handle different inputs | Specialize inherited behavior |

````python
# ── METHOD OVERLOADING EXAMPLE ────────────────────────────────────────────────

class Calculator:

    # ── Approach 1: Default Parameters ────────────────────────────
    def add(self, a, b=0, c=0, d=0):
        """Add 1 to 4 numbers"""
        result = a + b + c + d
        print(f"  add({', '.join(str(x) for x in [a,b,c,d] if x!=0)}) = {result}")
        return result

    # ── Approach 2: *args — any number of arguments ───────────────
    def multiply(self, *args):
        """Multiply any number of values"""
        result = 1
        for n in args:
            result *= n
        print(f"  multiply{args} = {result}")
        return result

    # ── Approach 3: Type checking inside method ───────────────────
    def area(self, *args):
        """
        Calculate area:
        - 1 arg  → circle (radius)
        - 2 args → rectangle (width, height)
        - 3 args → triangle (base, height, 'triangle')
        """
        import math
        if len(args) == 1:
            r = args[0]
            result = math.pi * r ** 2
            print(f"  Circle area (r={r}): {result:.2f}")
        elif len(args) == 2:
            w, h = args
            result = w * h
            print(f"  Rectangle area ({w}x{h}): {result:.2f}")
        elif len(args) == 3:
            b, h = args[0], args[1]
            result = 0.5 * b * h
            print(f"  Triangle area (b={b}, h={h}): {result:.2f}")
        else:
            print("  ❌ Invalid number of arguments.")
            result = None
        return result


calc = Calculator()

print("=== add() with different number of args ===")
calc.add(5)
calc.add(5, 10)
calc.add(5, 10, 20)
calc.add(5, 10, 20, 25)

print()
print("=== multiply() with *args ===")
calc.multiply(3, 4)
calc.multiply(2, 5, 10)
calc.multiply(1, 2, 3, 4, 5)

print()
print("=== area() with type checking ===")
calc.area(7)           # circle
calc.area(5, 10)       # rectangle
calc.area(6, 8, None)  # triangle
````

````python
# ── Why Python doesn't support traditional overloading ────────────────────────

class Demo:
    def greet(self):
        print("Hello!")

    def greet(self, name):         # This REPLACES the one above!
        print(f"Hello, {name}!")

d = Demo()
# d.greet()  # ❌ TypeError — only the second version exists now!
d.greet("Alice")  # ✅ works

print("Only the LAST definition of greet() survives in Python.")
print("Use default params or *args to simulate overloading!")

# ── Correct Pythonic way ───────────────────────────────────────────
class BetterDemo:
    def greet(self, name=None):          # default parameter
        if name:
            print(f"Hello, {name}!")
        else:
            print("Hello!")

bd = BetterDemo()
bd.greet()           # Hello!
bd.greet("Alice")    # Hello, Alice!
````

## 📌 Topic 11 — Operator Overloading

### 📖 Definition
**Operator Overloading** means redefining the behavior of a built-in
Python operator (`+`, `-`, `*`, `==`, `>`, `[]`, etc.) for your own
**custom class**. This allows objects of your class to behave naturally
with standard Python operators.

You implement this by defining special **dunder methods** (double underscore
methods) in your class.

### 🔑 Key Points
- Done using dunder (magic) methods like `__add__`, `__sub__`, `__eq__`
- Makes your custom objects feel like built-in types
- Python calls these methods automatically when operators are used
- Does NOT change operator behavior for built-in types

### 📋 Common Dunder Methods for Operators
| Operator | Dunder Method | Example |
|----------|--------------|---------|
| `+` | `__add__` | `v1 + v2` |
| `-` | `__sub__` | `v1 - v2` |
| `*` | `__mul__` | `v1 * 3` |
| `/` | `__truediv__` | `v1 / 2` |
| `==` | `__eq__` | `v1 == v2` |
| `!=` | `__ne__` | `v1 != v2` |
| `<` | `__lt__` | `v1 < v2` |
| `>` | `__gt__` | `v1 > v2` |
| `<=` | `__le__` | `v1 <= v2` |
| `>=` | `__ge__` | `v1 >= v2` |
| `len()` | `__len__` | `len(obj)` |
| `str()` | `__str__` | `print(obj)` |
| `repr()` | `__repr__` | `repr(obj)` |
| `[]` | `__getitem__` | `obj[0]` |
| `in` | `__contains__` | `x in obj` |

````python
# ── OPERATOR OVERLOADING EXAMPLE ──────────────────────────────────────────────

class Vector:
    """A 2D mathematical vector with operator overloading."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    # ── String representation ──────────────────────────────────────
    def __str__(self):               # for print()
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):              # for debugging/repr()
        return f"Vector(x={self.x}, y={self.y})"

    # ── Arithmetic operators ───────────────────────────────────────
    def __add__(self, other):        # v1 + v2
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):        # v1 - v2
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):       # v1 * 3  (scalar multiplication)
        return Vector(self.x * scalar, self.y * scalar)

    def __rmul__(self, scalar):      # 3 * v1  (reverse — scalar on left)
        return Vector(self.x * scalar, self.y * scalar)

    def __truediv__(self, scalar):   # v1 / 2
        return Vector(self.x / scalar, self.y / scalar)

    def __neg__(self):               # -v1 (unary negation)
        return Vector(-self.x, -self.y)

    # ── Comparison operators ───────────────────────────────────────
    def __eq__(self, other):         # v1 == v2
        return self.x == other.x and self.y == other.y

    def __lt__(self, other):         # v1 < v2 (compare magnitudes)
        return self.magnitude() < other.magnitude()

    def __le__(self, other):         # v1 <= v2
        return self.magnitude() <= other.magnitude()

    # ── Container-like behavior ────────────────────────────────────
    def __len__(self):               # len(v)
        return 2                     # always 2 components

    def __getitem__(self, index):    # v[0], v[1]
        return (self.x, self.y)[index]

    def __contains__(self, value):   # value in v
        return value == self.x or value == self.y

    # ── Helper method ──────────────────────────────────────────────
    def magnitude(self):
        return (self.x**2 + self.y**2) ** 0.5

    def dot(self, other):
        return self.x * other.x + self.y * other.y


v1 = Vector(3, 4)
v2 = Vector(1, 2)

print("=== Basic Operations ===")
print(f"v1         = {v1}")
print(f"v2         = {v2}")
print(f"v1 + v2    = {v1 + v2}")     # __add__
print(f"v1 - v2    = {v1 - v2}")     # __sub__
print(f"v1 * 3     = {v1 * 3}")      # __mul__
print(f"3 * v1     = {3 * v1}")      # __rmul__
print(f"v1 / 2     = {v1 / 2}")      # __truediv__
print(f"-v1        = {-v1}")         # __neg__

print()
print("=== Comparisons ===")
print(f"v1 == v2   = {v1 == v2}")    # __eq__
print(f"v1 == v1   = {v1 == v1}")
print(f"v1 < v2    = {v1 < v2}")     # __lt__ (by magnitude)
print(f"|v1|       = {v1.magnitude():.2f}")
print(f"|v2|       = {v2.magnitude():.2f}")

print()
print("=== Container-like ===")
print(f"len(v1)    = {len(v1)}")     # __len__
print(f"v1[0]      = {v1[0]}")       # __getitem__
print(f"v1[1]      = {v1[1]}")
print(f"3 in v1    = {3 in v1}")     # __contains__
print(f"9 in v1    = {9 in v1}")

# Sort vectors by magnitude — works because __lt__ is defined!
vectors = [Vector(5,5), Vector(1,1), Vector(3,4), Vector(2,2)]
vectors.sort()
print(f"\nSorted by magnitude: {[str(v) for v in vectors]}")
````

## 📌 Topic 12 — Abstraction

### 📖 Definition
**Abstraction** means **hiding complex internal details** and showing only
the **essential, relevant interface** to the user.
The user knows WHAT to do, but not HOW it's done internally.

In Python, abstraction is implemented using **Abstract Base Classes (ABC)**.
An abstract class defines methods that **must be implemented** by every
child class — if a child doesn't implement them, Python raises a `TypeError`
and you cannot even create the object.

### 🔑 Key Points
- Import: `from abc import ABC, abstractmethod`
- Decorate required methods with `@abstractmethod`
- Abstract class **cannot be instantiated** directly
- Any class that inherits must implement ALL abstract methods
- Can have both abstract AND concrete (normal) methods
- Enforces a **contract** — guarantees all subclasses have required methods

### 🧠 Abstraction vs Encapsulation
| | Abstraction | Encapsulation |
|--|-------------|---------------|
| **Focus** | Hide complexity of WHAT | Hide internal data of HOW |
| **Achieved by** | Abstract classes, interfaces | Private attributes, methods |
| **Purpose** | Design level — define contracts | Implementation level — protect data |
| **Example** | `abstract def area()` | `self.__balance = 0` |

### 🧠 Real-World Analogy
When you use a **smartphone**:
- You know: press home, swipe, tap app
- You do NOT know: ARM processor instructions, OS kernel calls, radio frequency handling
- That complexity is **abstracted away**. You only see the interface.

````python
# ── ABSTRACTION EXAMPLE ───────────────────────────────────────────────────────

from abc import ABC, abstractmethod

# ── ABSTRACT CLASS — the contract ─────────────────────────────────
class Notification(ABC):
    """
    Abstract base class for all notification types.
    Every subclass MUST implement send() and get_status().
    """

    def __init__(self, recipient, message):
        self.recipient = recipient
        self.message   = message
        self._sent     = False

    @abstractmethod
    def send(self):
        """Send the notification — HOW is up to each subclass"""
        pass

    @abstractmethod
    def get_status(self):
        """Return delivery status string"""
        pass

    # Concrete method — same for ALL subclasses (no override needed)
    def preview(self):
        print(f"  📋 Preview:")
        print(f"     To      : {self.recipient}")
        print(f"     Message : {self.message[:50]}...")
        print(f"     Via     : {self.__class__.__name__}")

    def log(self):
        status = self.get_status()     # calls overridden version!
        print(f"  📝 Log: [{self.__class__.__name__}] → {self.recipient} | {status}")


# ── CONCRETE CLASSES — implement the abstract methods ─────────────
class EmailNotification(Notification):

    def __init__(self, recipient, message, subject):
        super().__init__(recipient, message)
        self.subject = subject

    def send(self):                    # MUST implement
        print(f"  📧 Email sent!")
        print(f"     To      : {self.recipient}")
        print(f"     Subject : {self.subject}")
        print(f"     Body    : {self.message}")
        self._sent = True

    def get_status(self):              # MUST implement
        return "Delivered to inbox" if self._sent else "Pending"


class SMSNotification(Notification):

    def __init__(self, recipient, message, phone):
        super().__init__(recipient, message)
        self.phone = phone

    def send(self):
        print(f"  📱 SMS sent!")
        print(f"     Phone   : {self.phone}")
        print(f"     Message : {self.message[:160]}")  # SMS limit 160 chars
        self._sent = True

    def get_status(self):
        return "SMS delivered" if self._sent else "SMS pending"


class PushNotification(Notification):

    def __init__(self, recipient, message, device_token):
        super().__init__(recipient, message)
        self.device_token = device_token

    def send(self):
        print(f"  🔔 Push notification sent!")
        print(f"     Device  : {self.device_token[:20]}...")
        print(f"     Alert   : {self.message}")
        self._sent = True

    def get_status(self):
        return "Push delivered" if self._sent else "Push pending"


# ── Cannot instantiate abstract class ─────────────────────────────
print("=== Testing Abstract Class ===")
try:
    n = Notification("test@test.com", "hello")
except TypeError as e:
    print(f"  ❌ {e}")

print()
print("=== Sending Notifications ===")

notifications = [
    EmailNotification("alice@company.com",
                      "Your monthly report is ready. Please review.",
                      "Monthly Report - March 2024"),
    SMSNotification("Alice",
                    "OTP for login: 4821. Valid for 10 minutes.",
                    "+91-9876543210"),
    PushNotification("Alice",
                     "New message from Bob!",
                     "device_token_xyz_abc_123_456"),
]

for notif in notifications:
    print(f"\n── {notif.__class__.__name__} ──")
    notif.preview()        # concrete method — same for all
    notif.send()           # abstract method — different for each
    notif.log()            # concrete method using get_status()
````

## 📌 Topic 13 — Static Method

### 📖 Definition
A **static method** is a method that belongs to the **class** but does NOT
have access to the **instance** (`self`) or the **class** (`cls`).
It is just a regular function that is placed inside a class because it is
**logically related** to the class.

You define it using the `@staticmethod` decorator.

### 🔑 Key Points
- Decorator: `@staticmethod`
- No `self` or `cls` parameter
- Cannot access or modify instance variables or class variables
- Called on the class directly: `ClassName.method()` (preferred)
- Can also be called on an object: `obj.method()` (but not recommended)
- Used for **utility/helper functions** that relate to the class concept

### ⚙️ When to use Static Method?
Use a static method when the logic is related to the class but doesn't
need to know anything about the specific instance or class state.

### 🧠 Comparison: Instance vs Class vs Static Method
| | Instance Method | Class Method | Static Method |
|--|----------------|--------------|---------------|
| Decorator | none | `@classmethod` | `@staticmethod` |
| First param | `self` | `cls` | nothing |
| Access instance vars | ✅ Yes | ❌ No | ❌ No |
| Access class vars | ✅ Yes | ✅ Yes | ❌ No |
| Call on instance | ✅ | ✅ | ✅ |
| Call on class | ❌ | ✅ | ✅ |

````python
# ── STATIC METHOD EXAMPLE ─────────────────────────────────────────────────────

class StringUtils:
    """A utility class — all static methods, no instance needed."""

    @staticmethod
    def is_palindrome(s):
        """Check if string reads the same forwards and backwards."""
        cleaned = s.lower().replace(" ", "").replace(",", "").replace(".", "")
        return cleaned == cleaned[::-1]

    @staticmethod
    def is_anagram(s1, s2):
        """Check if two strings are anagrams of each other."""
        return sorted(s1.lower()) == sorted(s2.lower())

    @staticmethod
    def count_vowels(s):
        """Count vowels in a string."""
        return sum(1 for c in s.lower() if c in "aeiou")

    @staticmethod
    def title_case(s):
        """Capitalize first letter of each word."""
        return " ".join(word.capitalize() for word in s.split())

    @staticmethod
    def reverse_words(s):
        """Reverse the order of words in a string."""
        return " ".join(s.split()[::-1])

    @staticmethod
    def truncate(s, max_len=50, suffix="..."):
        """Shorten a string if it exceeds max_len."""
        if len(s) <= max_len:
            return s
        return s[:max_len - len(suffix)] + suffix


class MathUtils:

    @staticmethod
    def is_prime(n):
        if n < 2: return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0: return False
        return True

    @staticmethod
    def factorial(n):
        if n < 0: raise ValueError("Negative factorial undefined")
        result = 1
        for i in range(2, n+1):
            result *= i
        return result

    @staticmethod
    def gcd(a, b):
        """Greatest Common Divisor using Euclidean algorithm."""
        while b:
            a, b = b, a % b
        return a

    @staticmethod
    def celsius_to_fahrenheit(c):
        return (c * 9/5) + 32

    @staticmethod
    def is_perfect_square(n):
        root = int(n ** 0.5)
        return root * root == n


print("=== StringUtils (called on class — no object needed) ===")
print(f"  is_palindrome('racecar')        : {StringUtils.is_palindrome('racecar')}")
print(f"  is_palindrome('A man a plan a canal Panama') : {StringUtils.is_palindrome('A man a plan a canal Panama')}")
print(f"  is_anagram('listen','silent')   : {StringUtils.is_anagram('listen','silent')}")
print(f"  count_vowels('Hello World')     : {StringUtils.count_vowels('Hello World')}")
print(f"  title_case('hello world')       : {StringUtils.title_case('hello world')}")
print(f"  reverse_words('Python is fun')  : {StringUtils.reverse_words('Python is fun')}")
print(f"  truncate('This is a very long string that should be truncated') : {StringUtils.truncate('This is a very long string that should be truncated', 30)}")

print()
print("=== MathUtils ===")
primes = [n for n in range(2, 50) if MathUtils.is_prime(n)]
print(f"  Primes up to 50 : {primes}")
print(f"  factorial(10)   : {MathUtils.factorial(10)}")
print(f"  gcd(48, 18)     : {MathUtils.gcd(48, 18)}")
print(f"  100°C in °F     : {MathUtils.celsius_to_fahrenheit(100)}")
perfect_squares = [n for n in range(1, 50) if MathUtils.is_perfect_square(n)]
print(f"  Perfect squares : {perfect_squares}")
````

## 📌 Topic 14 — Class Method

### 📖 Definition
A **class method** is a method that receives the **class itself** as its
first argument (conventionally named `cls`) instead of the instance.
It is defined using the `@classmethod` decorator.

A class method has access to **class-level variables and methods** but
NOT to instance variables.

### 🔑 Key Points
- Decorator: `@classmethod`
- First parameter: `cls` (the class itself, not an object)
- Can read and modify **class variables**
- Cannot access `self` or instance variables
- Can be called on the **class** OR on an object
- Very useful for **factory methods** — alternative ways to create objects

### 🧠 Main Use Cases
**1. Access / modify class variables:**
Track how many objects have been created, change company name for all employees, etc.

**2. Factory methods (alternative constructors):**
Create objects from different input formats — from a string, from a file, from a dictionary, etc.
This is the most important use of class methods.

### ⚙️ Factory Method Pattern
```python
class Date:
    def __init__(self, day, month, year): ...

    @classmethod
    def from_string(cls, date_string):  # "15-03-2024"
        day, month, year = date_string.split("-")
        return cls(int(day), int(month), int(year))  # creates new object!
```

````python
# ── CLASS METHOD EXAMPLE ──────────────────────────────────────────────────────

class Employee:

    # Class variables
    company      = "IN2IT Technologies"
    total_count  = 0
    departments  = []

    def __init__(self, name, emp_id, department, salary):
        self.name       = name
        self.emp_id     = emp_id
        self.department = department
        self.salary     = salary
        Employee.total_count += 1
        if department not in Employee.departments:
            Employee.departments.append(department)

    def show(self):
        print(f"  [{self.emp_id}] {self.name:12s} | {self.department:15s} | ₹{self.salary:,}")

    # ── Class Method 1: Access class variable ─────────────────────
    @classmethod
    def get_company_info(cls):
        print(f"  🏢 Company   : {cls.company}")
        print(f"  👥 Employees : {cls.total_count}")
        print(f"  🏬 Depts     : {cls.departments}")

    # ── Class Method 2: Modify class variable ─────────────────────
    @classmethod
    def rename_company(cls, new_name):
        old = cls.company
        cls.company = new_name
        print(f"  ✅ Company renamed: {old} → {cls.company}")

    # ── Class Method 3: Factory method — from CSV string ──────────
    @classmethod
    def from_csv(cls, csv_string):
        """Create Employee from: 'name,id,department,salary'"""
        name, emp_id, dept, salary = csv_string.split(",")
        return cls(name.strip(), emp_id.strip(), dept.strip(), int(salary.strip()))

    # ── Class Method 4: Factory method — from dictionary ──────────
    @classmethod
    def from_dict(cls, data: dict):
        """Create Employee from a dictionary"""
        return cls(data["name"], data["id"], data["department"], data["salary"])

    # ── Class Method 5: Factory method — with default salary ──────
    @classmethod
    def create_intern(cls, name, emp_id, department):
        """Create an intern with fixed stipend of ₹15,000"""
        return cls(name, emp_id, department, 15000)


print("=== Creating Employees (regular way) ===")
e1 = Employee("Arjun",  "E001", "Engineering", 75000)
e2 = Employee("Priya",  "E002", "Marketing",   60000)

print()
Employee.get_company_info()

print()
print("=== Factory Methods — 3 different ways to create ===")

# From CSV string
e3 = Employee.from_csv("Rahul, E003, Finance, 55000")

# From dictionary
e4 = Employee.from_dict({
    "name": "Sneha",
    "id":   "E004",
    "department": "HR",
    "salary": 52000
})

# Intern factory
e5 = Employee.create_intern("Aditya", "I001", "Engineering")

print("All employees:")
for emp in [e1, e2, e3, e4, e5]:
    emp.show()

print()
Employee.get_company_info()

print()
print("=== Changing company name — affects ALL objects ===")
Employee.rename_company("Google India")
e1.show()   # shows new company? no — company is separate from show()
Employee.get_company_info()
````

## 📌 Topic 15 — Quick Summary & Interview Cheat Sheet

### 🗂️ All 14 Topics at a Glance

| # | Topic | One-Line Definition | Key Syntax |
|---|-------|---------------------|-----------|
| 1 | **Class** | Blueprint/template for creating objects | `class MyClass:` |
| 2 | **Object** | Real instance created from a class | `obj = MyClass()` |
| 3 | **Constructor** | Auto-called on object creation to initialize | `def __init__(self):` |
| 4 | **Destructor** | Auto-called on object destruction for cleanup | `def __del__(self):` |
| 5 | **Encapsulation** | Bundle data + methods, hide what's private | `self.__private` |
| 6 | **Getter/Setter** | Controlled read/write access to private data | `@property` |
| 7 | **Inheritance** | Child gets all parent's attributes + methods | `class Child(Parent):` |
| 8 | **Method Overriding** | Child redefines a parent method | Same name in child |
| 9 | **Polymorphism** | Same method name, different behavior per class | Loop + call same method |
| 10 | **Method Overloading** | Same name, handle different params (via defaults) | `def f(a, b=0):` |
| 11 | **Operator Overloading** | Redefine `+`, `-`, `==` for custom class | `def __add__(self, other):` |
| 12 | **Abstraction** | Hide complexity, enforce interface via ABC | `@abstractmethod` |
| 13 | **Static Method** | Utility function in class — no self/cls | `@staticmethod` |
| 14 | **Class Method** | Access/modify class vars or factory constructor | `@classmethod` |

---

### 🔑 The 4 OOP Pillars
```
┌─────────────────┬────────────────────────────────────────────────────────┐
│ Encapsulation   │ Hide data using __private, expose via methods          │
│ Inheritance     │ class Child(Parent) — reuse code, extend behavior      │
│ Polymorphism    │ Same method name, different result per class           │
│ Abstraction     │ Abstract class = WHAT to do, subclass = HOW to do it  │
└─────────────────┴────────────────────────────────────────────────────────┘
```
