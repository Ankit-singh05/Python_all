## Class

**Definition:**  
A *class* is a blueprint or template for creating objects. It defines the attributes (data) and methods (functions) that its objects will have.  
A class itself does nothing — it is simply the design.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print(f"{self.name} says: Woof!")
```

## Object

**Definition:**  
An *object* is a real instance created from a class. When you create an object, Python allocates memory and fills in the attributes using the class blueprint.  
You can create many objects from one class, each with its own data.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print(f"{self.name} says: Woof!")

# Creating two objects from the same class
dog1 = Dog("Bruno", "Labrador")
dog2 = Dog("Max", "Poodle")

dog1.bark()   # Bruno says: Woof!
dog2.bark()   # Max says: Woof!

print(dog1.name)   # Bruno
print(dog2.breed)  # Poodle
```
