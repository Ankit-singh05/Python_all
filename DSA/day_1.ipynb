# Array vs Linked List

|                | **Array**                                   | **Linked List**                                 |
|----------------|---------------------------------------------|-------------------------------------------------|
| **Type**       | Static data structure                       | Dynamic data structure                          |
| **Size**       | Fixed; cannot change at runtime              | Flexible; can change at runtime                 |
| **Memory**     | Contiguous memory allocation                 | Non-contiguous (scattered) memory allocation    |
| **Allocation** | Allocated once at creation                   | Allocated as needed during runtime              |

---
**Array**
- Memory allocation is done only once.
- Size is fixed and cannot be changed at runtime.
- Memory allocation is contiguous.

**Linked List**
- Memory allocation is done whenever required.
- Size is not fixed and can change at runtime.
- Memory allocation is non-contiguous (scattered).


## Array and Its Types

An **array** is a collection of elements, all of the same data type, stored in contiguous memory locations. Arrays allow efficient access to elements using an index.

### Types of Arrays

- **One-dimensional Array**: A simple list of elements, accessed by a single index.
  - Example: `int arr[5] = {1, 2, 3, 4, 5};`

- **Two-dimensional Array**: An array of arrays, often used to represent matrices or tables.
  - Example: `int matrix[3][3];`

- **Multi-dimensional Array**: Arrays with more than two dimensions, used for complex data structures.
  - Example: `int arr[2][3][4];`

Arrays can be of any data type, such as integers, floats, or characters.


## Linked List and Its Types

A **linked list** is a linear data structure where elements, called nodes, are stored in non-contiguous memory locations. Each node contains data and a reference (pointer) to the next node in the sequence.

### Types of Linked Lists

- **Singly Linked List**: Each node points to the next node. Traversal is possible in one direction only.
  - Example: `Node1 -> Node2 -> Node3 -> NULL`

- **Doubly Linked List**: Each node has two pointers, one to the next node and one to the previous node. Traversal is possible in both directions.
  - Example: `NULL <- Node1 <-> Node2 <-> Node3 -> NULL`

- **Circular Linked List**: The last node points back to the first node, forming a circle. Can be singly or doubly linked.
  - Example: `Node1 -> Node2 -> Node3 -> Node1`

Linked lists are useful for dynamic memory allocation and efficient insertions/deletions.
