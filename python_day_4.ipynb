## 📌 LESSON 5 — Functions

A function is a reusable, named block of code. Define once, call many times.
Functions help in **DRY** (Don't Repeat Yourself) programming.

```python
def greet(name,greeting = "Hello"):
    return f"{greeting} {name}"

print(greet("John"))
print(greet("Ankit","Hi"))
print(greet("John",greeting = "How are you"))
```

```python
# --- *args and **kwargs ---

def add(*args):
    print(f"args = {args}")
    return sum(args)

def show_info(**kwargs):
    print(f"kwargs = {kwargs}")
    for key , value in kwargs.items():
        print(f"{key} = {value}")
              
show_info(name="Alice", age=25, city="Delhi")
add(1,2,3,4,5)
```
