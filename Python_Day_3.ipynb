## 📌 LESSON 6 — Object-Oriented Programming (OOP)

OOP organizes code around **objects** which bundle data (attributes) and behavior (methods).

### The 4 Pillars of OOP
| Pillar | Meaning |
|--------|---------|
| **Encapsulation** | Hide internal data, expose only what's needed |
| **Inheritance** | Child class gets parent class's properties/methods |
| **Polymorphism** | Same method name, different behavior per class |
| **Abstraction** | Hide complexity, show only the interface |

# 🔒 PILLAR 1 — ENCAPSULATION

## What is Encapsulation?
Encapsulation means hiding the internal details of an object and only exposing what is necessary. You put a protective wall around the data so that outside code cannot directly access or mess with it. Access is controlled through specific methods.

> **Analogy:**  
> Think of it like a TV remote. You press the volume button — you don't need to know the circuit inside. The internal wiring is hidden. You only interact through the buttons (the interface).

## Encapsulation in Python

- **Public**: `name`  
  Anyone can access it from anywhere.

- **Protected**: `_name`  
  Convention meaning "don't touch this from outside, it's internal". Python doesn't enforce this, it's just a warning to other developers.

- **Private**: `__name`  
  Python applies name mangling. The attribute is renamed internally to `_ClassName__name`. This makes it harder (but not impossible) to access from outside.

```python
class BankAccount:

    def __init__(self, owner, balance):
        self.owner = owner          # public — anyone can access
        self._branch = "Noida"      # protected — internal use only
        self.__balance = balance    # private — strictly hidden

    # Controlled way to READ the balance
    def get_balance(self):
        return self.__balance

    # Controlled way to ADD money
    def deposit(self, amount):
        if amount <= 0:
            print("Amount must be positive.")
            return
        self.__balance += amount
        print(f"Deposited ₹{amount}. New balance: ₹{self.__balance}")

    # Controlled way to REMOVE money — with validation
    def withdraw(self, amount):
        if amount <= 0:
            print("Amount must be positive.")
            return
        if amount > self.__balance:
            print("Insufficient funds!")
            return
        self.__balance -= amount
        print(f"Withdrawn ₹{amount}. Remaining: ₹{self.__balance}")


# Create an account
account = BankAccount("Alice", 10000)

# ✅ This works — public attribute
print(account.owner)

# ✅ This works — through the controlled method
account.deposit(5000)
account.withdraw(3000)
print(account.get_balance())

# ❌ This would FAIL — private attribute is hidden
# print(account.__balance)  → AttributeError

# ⚠️ This technically works but is bad practice — never do this
print(account._BankAccount__balance)  # name mangling bypass
```

```python
# Encapsulation Example
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def display(self):
    print("Name:", self.name)
    print("Age:", self.age)

person = Person("John", 30)
person.display()
```

# 🟥 PILLAR 2 — INHERITANCE

**What it means in simple words:**

Inheritance means a child class automatically gets all the properties and methods of the parent class. You don't have to rewrite the same code again. The child can then add its own new things or change (override) existing things.

Think of it like a biological family. You inherit your parents' traits — eye color, height tendencies — but you are also your own person with your own unique traits.

In programming terms: write common code once in the parent, reuse it in all children.

```python
# PARENT CLASS — contains common behavior
class Vehicle:

    def __init__(self, brand, speed, fuel):
        self.brand = brand
        self.speed = speed
        self.fuel  = fuel

    def start(self):
        print(f"{self.brand} is starting...")

    def stop(self):
        print(f"{self.brand} has stopped.")

    def info(self):
        print(f"Brand: {self.brand} | Speed: {self.speed} km/h | Fuel: {self.fuel}")


# CHILD CLASS 1 — inherits everything from Vehicle, adds its own things
class Car(Vehicle):

    def __init__(self, brand, speed, fuel, doors):
        super().__init__(brand, speed, fuel)  # call parent constructor
        self.doors = doors                    # Car's own attribute

    def air_conditioning(self):               # Car's own method
        print(f"{self.brand} AC is ON.")

    def info(self):                           # OVERRIDE parent's info
        super().info()                        # still call parent's version
        print(f"Doors: {self.doors}")


# CHILD CLASS 2 — inherits from Vehicle, different kind
class Motorcycle(Vehicle):

    def __init__(self, brand, speed, fuel, has_sidecar):
        super().__init__(brand, speed, fuel)
        self.has_sidecar = has_sidecar

    def wheelie(self):
        print(f"{self.brand} does a wheelie! 🏍️")

    def info(self):
        super().info()
        print(f"Sidecar: {'Yes' if self.has_sidecar else 'No'}")


# CHILD CLASS 3 — inherits from Vehicle
class ElectricCar(Car):   # inherits from Car which inherits from Vehicle
                           # this is MULTILEVEL inheritance

    def __init__(self, brand, speed, doors, battery_range):
        super().__init__(brand, speed, "Electric", doors)
        self.battery_range = battery_range

    def charge(self):
        print(f"{self.brand} is charging... 🔋")

    def info(self):
        super().info()
        print(f"Battery Range: {self.battery_range} km")


# Using the classes
car   = Car("Toyota", 180, "Petrol", 4)
bike  = Motorcycle("Royal Enfield", 140, "Petrol", False)
tesla = ElectricCar("Tesla", 250, 4, 600)

car.start()        # inherited from Vehicle
car.air_conditioning()  # Car's own method
car.info()         # overridden version

print()
bike.start()       # inherited from Vehicle
bike.wheelie()     # Motorcycle's own method
bike.info()

print()
tesla.start()      # inherited all the way from Vehicle
tesla.charge()     # ElectricCar's own method
tesla.info()       # overridden

# Check relationships
print(isinstance(tesla, ElectricCar))  # True
print(isinstance(tesla, Car))          # True  — it IS a Car
print(isinstance(tesla, Vehicle))      # True  — it IS a Vehicle
print(isinstance(bike, Car))           # False — Motorcycle is not a Car
```

```python
# Inheritence Example
class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    pass

class Dog(Animal):
  def speak(self):
    return "Woof!"

class Cat(Animal):
  def speak(self):
    return "Meow!"

dog = Dog("Fido")
cat = Cat("Whiskers")
print(dog.speak())  # Output: Woof!
print(cat.speak())  # Output: Meow!
```

# 🟥 PILLAR 3 — POLYMORPHISM

**What it means in simple words:**

Polymorphism means same method name, different behavior depending on which class is calling it. The word itself comes from Greek — poly means many, morphism means forms. So the same method takes many forms.

Think of the word "speak". You tell a dog to speak and it barks. You tell a cat to speak and it meows. You tell a person to speak and they talk. Same instruction, completely different result based on who receives it.

**There are two types in Python:**

- **Method Overriding** — child class redefines a method from parent class. This is runtime polymorphism.
- **Duck Typing** — Python's own style of polymorphism. If an object has the method you're calling, Python will call it — it doesn't care what type the object is. "If it walks like a duck and quacks like a duck, it's a duck."

```python
# Base class
class Shape:

    def __init__(self, color):
        self.color = color

    def area(self):
        # This is meant to be overridden — raise error if not
        raise NotImplementedError("Every shape must define its own area()")

    def describe(self):
        # This method uses area() which will be different per shape
        # This is polymorphism — describe() works for ALL shapes
        # without knowing which specific shape it is
        return f"I am a {self.color} {self.__class__.__name__} with area {self.area():.2f} sq units"


class Circle(Shape):

    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def area(self):                          # OVERRIDE
        return 3.14159 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14159 * self.radius


class Rectangle(Shape):

    def __init__(self, color, width, height):
        super().__init__(color)
        self.width  = width
        self.height = height

    def area(self):                          # OVERRIDE
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


class Triangle(Shape):

    def __init__(self, color, base, height):
        super().__init__(color)
        self.base   = base
        self.height = height

    def area(self):                          # OVERRIDE
        return 0.5 * self.base * self.height


class Square(Rectangle):                    # inherits from Rectangle

    def __init__(self, color, side):
        super().__init__(color, side, side)  # width = height = side


# ── POLYMORPHISM IN ACTION ────────────────────────────────────────
shapes = [
    Circle("red", 7),
    Rectangle("blue", 5, 10),
    Triangle("green", 6, 8),
    Square("yellow", 4),
]

print("=== All shapes using the SAME describe() method ===")
for shape in shapes:
    print(shape.describe())   # same call → completely different result!

print()

# Python doesn't care what TYPE shape is — it just calls area()
# This is DUCK TYPING — if it has area(), call it
print("=== Total area of all shapes ===")
total = sum(shape.area() for shape in shapes)
print(f"Total area: {total:.2f} sq units")

print()

# Even a completely unrelated class works — duck typing!
class LandPlot:       # does NOT inherit from Shape at all!
    def area(self):
        return 5000.0

plot = LandPlot()
all_areas = shapes + [plot]    # mix Shape objects with LandPlot

# Python just calls .area() on whatever it finds — no type check!
print("Total including land plot:", sum(x.area() for x in all_areas))
```

```python
# Polymorphism Example
class Shape:
  def area(self):
    pass

class Rectangle(Shape):
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius ** 2

shapes = [Rectangle(5, 10), Circle(7)]
for shape in shapes:
  print(shape.area())
```


# 🎭 PILLAR 4 — ABSTRACTION

## What it means in simple words

Abstraction means hiding the complex internal implementation and showing only what the user needs to know. You define **WHAT** something should do, but not **HOW** it does it. The HOW is hidden inside.

> **Analogy:**  
> Think of driving a car. You know: turn key to start, press accelerator to go, press brake to stop.  
> You do NOT need to know how the fuel injection works, how the pistons move, or how the brake pads function.  
> That complexity is abstracted away. You only see the simple interface.

In Python, abstraction is implemented using **Abstract Base Classes (ABC)**.  
An abstract class defines methods that MUST be implemented by every child class.  
If a child doesn't implement them, Python raises an error — you cannot even create the object.

```python
from abc import ABC, abstractmethod

# ABSTRACT CLASS — blueprint, cannot be instantiated directly
class PaymentGateway(ABC):

    def __init__(self, merchant_name):
        self.merchant_name = merchant_name

    # Abstract methods — MUST be implemented by every child class
    # These define WHAT the gateway must do — not HOW
    @abstractmethod
    def connect(self):
        pass

    @abstractmethod
    def process_payment(self, amount, currency):
        pass

    @abstractmethod
    def refund(self, transaction_id, amount):
        pass

    # Concrete method — already implemented, shared by all children
    # Children inherit this as-is
    def show_receipt(self, amount, currency, transaction_id):
        print(f"""
        ─────────────────────────────
        ✅ PAYMENT SUCCESSFUL
        Merchant   : {self.merchant_name}
        Amount     : {amount} {currency}
        TxnID      : {transaction_id}
        Gateway    : {self.__class__.__name__}
        ─────────────────────────────""")


# CONCRETE CLASS 1 — implements all abstract methods in its own way
class RazorpayGateway(PaymentGateway):

    def connect(self):
        print("Razorpay: Connecting via REST API with API key...")
        print("Razorpay: Secure connection established ✓")

    def process_payment(self, amount, currency):
        print(f"Razorpay: Charging {amount} {currency} via UPI/Card...")
        print("Razorpay: Verifying with bank...")
        txn_id = "RZP" + str(amount) + "TXN"
        print("Razorpay: Payment approved!")
        self.show_receipt(amount, currency, txn_id)
        return txn_id

    def refund(self, transaction_id, amount):
        print(f"Razorpay: Initiating refund of ₹{amount} for {transaction_id}")
        print("Razorpay: Refund will reflect in 5-7 business days.")


# CONCRETE CLASS 2 — completely different implementation
class PayPalGateway(PaymentGateway):

    def connect(self):
        print("PayPal: Authenticating with OAuth 2.0...")
        print("PayPal: Fetching access token...")
        print("PayPal: Ready ✓")

    def process_payment(self, amount, currency):
        print(f"PayPal: Processing {amount} {currency} via PayPal wallet...")
        print("PayPal: Checking fraud filters...")
        txn_id = "PPL" + str(amount) + "TXN"
        print("PayPal: Transaction complete!")
        self.show_receipt(amount, currency, txn_id)
        return txn_id

    def refund(self, transaction_id, amount):
        print(f"PayPal: Reversing transaction {transaction_id}")
        print(f"PayPal: ₹{amount} will be returned immediately.")


# CONCRETE CLASS 3
class StripeGateway(PaymentGateway):

    def connect(self):
        print("Stripe: Initializing SDK...")
        print("Stripe: Webhook endpoint registered ✓")

    def process_payment(self, amount, currency):
        print(f"Stripe: Creating payment intent for {amount} {currency}...")
        print("Stripe: Confirming with 3D Secure...")
        txn_id = "STR" + str(amount) + "TXN"
        print("Stripe: Payment confirmed!")
        self.show_receipt(amount, currency, txn_id)
        return txn_id

    def refund(self, transaction_id, amount):
        print(f"Stripe: Creating refund object for {transaction_id}")
        print(f"Stripe: ₹{amount} queued for refund.")


# ── ABSTRACTION IN ACTION ──────────────────────────────────────────

# This would FAIL — abstract class cannot be instantiated
# gateway = PaymentGateway("MyShop")  → TypeError!

# ── The power of abstraction: ─────────────────────────────────────
# This function works with ANY payment gateway
# It doesn't know or care whether it's Razorpay, PayPal, or Stripe
# It only knows: connect(), process_payment(), refund() exist

def checkout(gateway: PaymentGateway, amount, currency="INR"):
    print(f"\n{'='*45}")
    print(f"Starting checkout via {gateway.__class__.__name__}")
    print(f"{'='*45}")
    gateway.connect()
    txn = gateway.process_payment(amount, currency)
    return txn


# Use all three gateways through the SAME interface
razorpay = RazorpayGateway("IN2IT Store")
paypal   = PayPalGateway("IN2IT Store")
stripe   = StripeGateway("IN2IT Store")

txn1 = checkout(razorpay, 1999)
txn2 = checkout(paypal,   4999, "USD")
txn3 = checkout(stripe,   2499)

# Refund using same abstract interface
print("\n=== Processing Refunds ===")
razorpay.refund(txn1, 1999)
paypal.refund(txn2, 4999)
```

```python
# Abstraction Example
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Rectangle(Shape):
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius ** 2

shapes = [Rectangle(5, 10), Circle(7)]
for shape in shapes:
  print(shape.area())
```
