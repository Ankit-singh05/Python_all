 ***LESSON 1: What is Python?***

 Python is a high-level, interpreted programming language. That means you write simple English-like code and the computer understands it directly — no need to compile like C or Java.

 ***Why Python is popular:***
- Very easy to read and write  
- Works for web development, data science, AI, automation  
- Huge collection of ready-made libraries  
- Used by Google, Instagram, Netflix, NASA

- **Python Data Types and Variables**
  - Python has five standard data types:
    - Numbers
    - String
    - List
    - Tuple
    - Dictionary
  - Python has no command for declaring a variable. A variable is created the moment you first assign a value to it.

```python
#Example Of Data type 
x = 5
print(type(x)) # <class 'int'>

y = "Hello"
print(type(y)) # <class 'str'>

z = 3.14
print(type(z)) # <class 'float'>

a = True
print(type(a)) # <class 'bool'>

b = [1, 2, 3]
print(type(b)) # <class 'list'>

c = {"name": "John", "age": 30}
print(type(c)) # <class 'dict'>

d = (1, 2, 3)
print(type(d)) # <class 'tuple'>

e = {1, 2, 3}
print(type(e)) # <class 'set'>
```

```python
# --- Multiple Assignment & Swapping ---
x = y = z = 100          # all point to same value
a, b, c = 1, 2, 3        # tuple unpacking
print(a, b, c)

a, b = b, a              # swap without temp variable!
print(a, b)
```

```python
#String

st = "Hello"
s = "  Hello World  "

print(s.strip())#[Remove White Space]
print(s.upper())
print(s.lower())
print(s.strip().split(" "))
print(s.replace("Hello","Python"))
```

```python
# --- String Slicing ---
s = "Python"

print(s[:1])
print(s[1:])
print(s[::-1])  # 'nohtyP' → REVERSED (very common in interviews!)
print(s[::2])   # 'Pto' → every 2nd character
```

```python
# ====== LIST ======
lst = [3, 1, 4, 1, 5, 9, 2, 6]

lst.append(3)          # [3, 1, 4, 1, 5, 9, 2, 6, 3]
lst.sort()              # [1, 1, 2, 3, 3, 4, 5, 6, 9]
lst.reverse()           # [9, 6, 5, 4, 3, 3, 2, 1, 1]
lst.remove(3)           # [9, 6, 5, 4, 3, 2, 1, 1]
lst.pop()               # [9, 6, 5, 4, 3, 2, 1]
```
