**Questions**

1. Find whether a number is prime.
2. Determine if a number is even or odd.
3. Reverse a string without using slicing.
4. Check if a number is a palindrome.
5. Remove duplicates from a list.

```python
def prime(n):
        if all(n%j !=0 for j in range(2,int(n**0.5)+1) ):
            print(f"it is prime no.:- {n}")
        else:
            print(f"it is not prime no.:- {n}")
    
def even(n):
        if n % 2 == 0 :
            print(f"it is even no.:- {n}")
        else:
            print(f"it is odd no.:- {n}")

def remove_duplicate(n):
    print(list(set(n)))
    
def string_reverse(n):
    em = ""
    for i in n:
        em = i + em
    print(em)

def palindrom(n):
    if str(n) == str(n)[::-1]:
        print(f"{n} is a palindrome")
    else:
        print(f"{n} is not palindrome")

nu = 142
lst = [2,4,5,6,7,1,2,3,4,5,2,3,5]
st = "Ankit"
palindrom(nu)
string_reverse(st)  
remove_duplicate(lst)
even(5)
prime(6)
```

**Questions**
1. Pattern of square box
2. Pattern of pyramid
3. Pattern of pyramid with increasing numbers
4. Pattern of right angle triangle

```python
# Pattern of square box
# Pattern of pyramid
# Pattern of pyramid with increasing numbers
# Pattern of right angle triangle

def square_box_full(n):
    for i in range(n):
        print("* " * n)
        
def square_box_empty(n):
    for i in range(n):
        if i == 0 or i == n-1:
            print("* " * n)
        else:
            print("* " + "  " * (n-2) + "* ")

def pyramid(n):
    for i in range(1,n+1):
        print(" " * (n-i) + "* " * i)

def pyramid_inc_no(n):
    for i in range(1,n+1):
        print(" " * (n-i) + (str(i) + " ") * i)
        
def right_angle(n):
    for i in range(n):
        print("* " * i)
# square_box_full(5)
# square_box_empty(6)
# pyramid(5)
```

**Question**
1. list comprehensive 
2. lambda function 
3. function
4. what is argument and key arguments 
5. map and filter 

```python
# List Comprehension
# Dictionary Comprehension
# Set Comprehension
# Generator Comprehension

def list_comp(n):
    return [i for i in range(n)]
def dict_comp(n):
    return {i:i for i in range(n)}
def set_comp(n):
    return {i for i in range(n)}
def gen_comp(n):
    return (i for i in range(n))
print(list_comp(5))
print(dict_comp(5))
print(set_comp(5))
print(gen_comp(5))

# lambda function 
st = lambda x,y:x+y 
print(st(2,3))
```
