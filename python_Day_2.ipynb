```python
# --- Dictionary ---
d = {"name": "John", "age": 30}

print(d["name"]) # John
print(d.get("age")) # 30
d["age"] = 31
d["city"] = "New York"
print(d) # {'name': 'John', 'age': 31, 'city': 'New York'}
d.pop("city")
print(d) # {'name': 'John', 'age': 31}
d.update({"name": "Jane", "city": "New York"})
print(d) # {'name': 'Jane', 'age': 31, 'city': 'New York'}
d.clear()
print(d) # {}
# --- Set ---
s = {1, 2, 3, 4, 5}  # Set
s.add(6)
s.remove(3)
print(s) # {1, 2, 4, 5, 6}
s2 = {3, 4, 5, 6, 7}
print(s.union(s2)) # {1, 2, 3, 4, 5, 6, 7}
print(s.intersection(s2)) # {3, 4, 5, 6}
print(s.difference(s2)) # {1, 2}
print(s.symmetric_difference(s2)) # {1, 2, 7}
# --- Tuple ---
t = (1, 2, 3)
print(t[0]) # 1
print(t[1:]) # (2, 3)
print(t[::-1]) # (3, 2, 1)
print(t[::2]) # (1, 3)
# --- Boolean ---
b = True
print(not b) # False
print(b and False) # False
print(b or False) # True
print(b and True) # True
print(b or True) # True
print(b and b) # True
print(b or b) # True
print(b and not b) # False
print(b or not b) # True
print(b and b and b) # True
print(b or b or b) # True
print(b and b or b) # True
print(b or b and b) # True
print(b and b and b and b) #    True
print(b or b or b or b) #    True
print(b and b or b or b) #    True
print(b or b and b or b) #    True
print(b and b and b and b and b) #    True
print(b or b or b or b or b) #    True
print(b and b or b or b or b)
```
