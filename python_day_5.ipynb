## 📌 LESSON 7 — Decorators

A decorator is a function that **wraps another function** to add extra behavior — without changing the original function's code.

Think of it like a gift wrapper — the gift (function) doesn't change, but it gets extra packaging.


Think of a decorator like gift wrapping for a function.
The function (the gift) stays the same inside, but the decorator (the wrapping) adds something extra on the outside—like a ribbon or a card—without changing what's inside.

**Simple Example: The "Greeting" Decorator**

Imagine you have a function that just prints a name. You want to add "Hello" before it and "Nice to meet you" after, without modifying the original function's code.

```python
# 1. This is the Decorator (the "Wrapping")
def my_decorator(func):
    def wrapper():
        print("Hello!")            # Added behavior before
        func()                     # The original function runs here
        print("Nice to meet you.") # Added behavior after
    return wrapper

# 2. This is the original function (the "Gift")
@my_decorator
def say_name():
    print("I am Python.")

# 3. Run it
say_name()
```

```python
import time

# 1. The Decorator
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()          # Record start time
        result = func(*args, **kwargs)   # Run the actual function
        end_time = time.time()            # Record end time
        
        duration = end_time - start_time
        print(f"Function '{func.__name__}' took {duration:.4f} seconds.")
        return result
    return wrapper

# 2. Apply it to a "heavy" function
@timer_decorator
def heavy_task():
    # This simulates a task that takes time
    sum(i**2 for i in range(1000000))

heavy_task()
```

```python
# --- How a Decorator Works (step by step) ---

def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("▶ Before the function runs")
        result = func(*args, **kwargs)   # call original function
        print("◀ After the function runs")
        return result
    return wrapper

# Apply decorator using @ syntax
@my_decorator
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
# Output:
# ▶ Before the function runs
# Hello, Alice!
# ◀ After the function runs
```

```python
# Decorator function
def decorator_function(func):
    def wrapper_function(*args,**kwargs):
        print("Before function execution")
        func(*args,**kwargs)
        print("After function execution")
    return wrapper_function

@decorator_function
def my_function():
    print("Inside my_function")
my_function()
@decorator_function
def my_function_with_args(arg1,arg2):
    print(f"Inside my_function_with_args with arguments: {arg1}, {arg2}")
my_function_with_args("Hello", "World")
```

```python
# Decorator
# Decorator is a function which takes another function as an argument and extend the behaviour of the function without explicitly modifying it.

def decorator(func):
    def wrapper(*args,**kwargs):
        print("Before function call")
        func(*args,**kwargs)
        print("After function call")
    return wrapper

@decorator
def func1():
    print("This is function 1")

@decorator
def func2(a,b):
    print("This is function 2")
    print(a+b)

func1()
func2(2,3)

```
