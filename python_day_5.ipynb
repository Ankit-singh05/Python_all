## 📌 LESSON 7 — Decorators

A decorator is a function that **wraps another function** to add extra behavior — without changing the original function's code.

Think of it like a gift wrapper — the gift (function) doesn't change, but it gets extra packaging.


Think of a decorator like gift wrapping for a function.
The function (the gift) stays the same inside, but the decorator (the wrapping) adds something extra on the outside—like a ribbon or a card—without changing what's inside.

**Simple Example: The "Greeting" Decorator**

Imagine you have a function that just prints a name. You want to add "Hello" before it and "Nice to meet you" after, without modifying the original function's code.

```python
# 1. This is the Decorator (the "Wrapping")
def my_decorator(func):
    def wrapper():
        print("Hello!")            # Added behavior before
        func()                     # The original function runs here
        print("Nice to meet you.") # Added behavior after
    return wrapper

# 2. This is the original function (the "Gift")
@my_decorator
def say_name():
    print("I am Python.")

# 3. Run it
say_name()
```

1. What are *args and **kwargs?

**`*args` (Non-keyword Arguments):**  
The asterisk (`*`) is used to collect additional positional arguments into a tuple.  
*Example:* Passing `1, 2, 3` results in `args` being `(1, 2, 3)`.

**`**kwargs` (Keyword Arguments):**  
The double asterisk (`**`) collects extra named (keyword) arguments into a dictionary.  
*Example:* Passing `age=25` results in `kwargs` being `{'age': 25}`.

```python
import time

# 1. The Decorator
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()          # Record start time
        result = func(*args, **kwargs)   # Run the actual function
        end_time = time.time()            # Record end time
        
        duration = end_time - start_time
        print(f"Function '{func.__name__}' took {duration:.4f} seconds.")
        return result
    return wrapper

# 2. Apply it to a "heavy" function
@timer_decorator
def heavy_task():
    # This simulates a task that takes time
    sum(i**2 for i in range(1000000))

heavy_task()
```

```python
# --- How a Decorator Works (step by step) ---

def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("▶ Before the function runs")
        result = func(*args, **kwargs)   # call original function
        print("◀ After the function runs")
        return result
    return wrapper

# Apply decorator using @ syntax
@my_decorator
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
# Output:
# ▶ Before the function runs
# Hello, Alice!
# ◀ After the function runs
```

✨ **Decorators can also be used to measure the execution time of a function.**

- `{func.__name__}`: Automatically retrieves the name of the function being timed.

This makes it easy to see how long your functions take to run, without changing their original code!

### Why do we use `.4f`?

Without formatting, Python might display a long, messy number like `0.12456789123`.  
By using `.4f`, you make it clean and readable.

---

#### Simple Examples

| Format           | Result (for `3.14159`) | Explanation                        |
|------------------|------------------------|------------------------------------|
| `{duration:.2f}` | `3.14`                 | Shows 2 decimal places             |
| `{duration:.0f}` | `3`                    | Rounds to the nearest whole number |
| `{duration:.5f}` | `3.14159`              | Shows 5 decimal places             |

```python
import time

def mydecorator(func):
    def wrapper(*args,**kargs):
        start = time.time()
        result = func(*args,**kargs)
        end = time.time()

        time_taken = end - start
        print(f" {func.__name__} took {time_taken:.4f} seconds to execute")
        return result
    return wrapper

@mydecorator
def sum_all(*args):
    return sum(args)

print(sum_all(1,2,3,4,5))
```

### 🌀 Generators

A **Generator** is a special type of function that returns an iterator—producing a sequence of values one at a time using the `yield` keyword instead of `return`.

**Why use generators?**  
- They save memory: Instead of creating the whole list at once, a generator produces the next item only when you ask for it.

**Simple Example:**

```python
def count_to_three():
    yield 1
    yield 2
    yield 3

# Usage
for num in count_to_three():
    print(num)
```

### 2️⃣ Closures

A **Closure** is an inner function that "remembers" and has access to variables from its outer function's scope—even after the outer function has finished executing.

**Why use closures?**  
- They let you "attach" data to a function without using global variables or classes.

**Simple Example:**
---

```python
def make_multiplier(x):
    # This is the outer function
    def multiply(n):
        # This inner function "remembers" x
        return n * x
    return multiply

# Usage
double = make_multiplier(2) # x is now 2
print(double(5)) # Output: 10 (5 * 2)

triple = make_multiplier(3) # x is now 3
print(triple(5)) # Output: 15 (5 * 3)
```
